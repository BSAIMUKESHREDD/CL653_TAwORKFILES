**Outlier Detection and removal**

**Z-score**

```python
"""this is a function to reMove otliers using zscore """
from scipy import stats

def remove_outliers_zscore(column):
    z_scores = stats.zscore(column)
    outlier_indices = abs(z_scores) > threshold
    column[outlier_indices] = column.mean()
    return column
```

IQR

```python
def remove_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column >= lower_bound) & (column <= upper_bound)]
```

```python
threshold=3  # change this value as u want
# Apply the function to all numeric columns in the DataFrame
df_no_outliers = data_df.apply(remove_outliers_zscore, axis=0)
#check after outlier detection and removal some times while dropping or replacing nan values may be created.
#change to iqr code and check urself
df_no_outliers.isna().sum()
```

```python
from sklearn.decomposition import PCA

num_components = 2  # change to 3 also and try
pca = PCA(n_components=num_components)
X_pca = pca.fit_transform(df_scaled)

# Create a new DataFrame with the principal components
columns_pca = [f'PC{i+1}' for i in range(num_components)]
df_pca = pd.DataFrame(data=X_pca, columns=columns_pca)
```

```python
transformation_matrix = pca.components_
# Display the transformation matrix
print("Transformation Matrix (Coefficients for New Equations):")
print(pd.DataFrame(transformation_matrix, columns=df_scaled.columns))
```

```python
num_components = X_pca.shape[1]
print("The equations for PC1 and PC2 are as follows")
for i in range(num_components):
    equation = f"PC{i + 1} = "
    for j in range(len(df_scaled.columns)):
        coefficient = transformation_matrix[i, j]
        feature_name = df_scaled.columns[j]
        equation += f"{coefficient:.3f} * {feature_name}"
        if j < len(df_scaled.columns) - 1:
            equation += " + "
    print(equation)
```
